```python
text = """Natural Language Processing (NLP) is a field of Artificial Intelligence.
It helps computers understand, interpret and generate human language.
NLP is widely used in chatbots, translation systems and sentiment analysis."""
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag

from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Download resource

nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')

nltk.download('averaged_perceptron_tagger_eng')
```

TOKENIZATION : Splits text into words


```python
tokens = word_tokenize(text)
print("Tokens:\n", tokens)
```

POS Tagging : Assigns grammatical tags like:
NN → Noun
VB → Verb
JJ → Adjective

```python
from nltk import pos_tag
pos_tags = pos_tag(tokens)
print(pos_tags)
```

Stop Words Removal : Removes common words like is, the, and

```python
stop_words = set(stopwords.words('english'))

filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
print("\nAfter Stopword Removal:\n", filtered_tokens)
```

Stemming : Reduces words to root (may not be real words)
Example: processing → process

```python
stemmer = PorterStemmer()

stemmed_words = [stemmer.stem(word) for word in filtered_tokens]
print( stemmed_words)
```

Lemmatization : Converts to meaningful base form
Example: better → good

```python
lemmatizer = WordNetLemmatizer()

lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_tokens]
print(lemmatized_words)
```

TF (Term Frequency) Representation : Counts frequency of each word

```python
from collections import Counter

tf = Counter(lemmatized_words)
print(tf)
```

TF-IDF Representation : Weights words based on importance:

High TF + Low DF → Important word
And 
Common words get lower scores


```python
documents = [text]

vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(documents)

print("\nFeature Names:\n", vectorizer.get_feature_names_out())
print("\nTF-IDF Matrix:\n", tfidf_matrix.toarray())
```
